```python
df = pd.read_csv('data/Electric_Vehicle_Population_Data.csv')


county_frequency = df['County'].value_counts()



top_7_counties = county_frequency.index[:7]

# Create a dictionary to store dataframes for each county
county_tables = {county: df[df['County'] == county] for county in top_7_counties}

# Connect to SQLite database (or create one if it doesn't exist)
conn = sqlite3.connect('data/electric_vehicle_data.db')

# Iterate through the county_tables dictionary and create tables
for county, table in county_tables.items():
    table.to_sql(county, conn, if_exists='replace', index=False)

# Close the connection
conn.close()
```
